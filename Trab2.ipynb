```python
import cv2
import os
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import LabelEncoder
from sklearn.neural_network import MLPClassifier
import numpy as np

# Defina o caminho para os diretórios onde estão armazenadas as imagens
diretorio_carro = 'C:\\Users\\Pedro\\Downloads\\AMMC\\carro'
diretorio_moto = 'C:\\Users\\Pedro\\Downloads\\AMMC\\motos'
diretorio_pedestre = 'C:\\Users\\Pedro\\Downloads\\AMMC\\pedestre'
diretorio_faixa = 'C:\\Users\\Pedro\\Downloads\\AMMC\\faixa'


# Função para carregar e redimensionar imagens de um diretório e rotular
def carregar_imagens(diretorio, rotulo, tamanho=(100, 100)):
    imagens = []
    rotulos = []
    for nome_arquivo in os.listdir(diretorio):
        caminho_arquivo = os.path.join(diretorio, nome_arquivo)
        imagem = cv2.imread(caminho_arquivo)
        imagem_redimensionada = cv2.resize(imagem, tamanho)  # Redimensionar imagem
        imagens.append(imagem_redimensionada)
        rotulos.append(rotulo)
    return imagens, rotulos

# Carregar imagens e rotular
imagens_carro, rotulos_carro = carregar_imagens(diretorio_carro, 'carro')
imagens_moto, rotulos_moto = carregar_imagens(diretorio_moto, 'moto')
imagens_pedestre, rotulos_pedestre = carregar_imagens(diretorio_pedestre, 'pedestre')
imagens_faixa, rotulos_faixa = carregar_imagens(diretorio_faixa, 'faixa')

# Concatenar todas as imagens e rótulos em uma lista
imagens = np.concatenate((imagens_carro, imagens_moto, imagens_pedestre, imagens_faixa), axis=0)
rotulos = np.concatenate((rotulos_carro, rotulos_moto, rotulos_pedestre, rotulos_faixa), axis=0)

# Converter os rótulos para o formato numérico
encoder = LabelEncoder()
y_encoded = encoder.fit_transform(rotulos)

# Achatando as imagens para que possam ser usadas como entrada
X_flat = np.array([imagem.flatten() for imagem in imagens])

# Dividir os dados em conjuntos de treinamento e teste
X_train, X_test, y_train, y_test = train_test_split(X_flat, y_encoded, test_size=0.2, random_state=42)

# Criar o classificador de rede neural
clf = MLPClassifier(hidden_layer_sizes=(100, 100), max_iter=1000)

# Treinar o classificador
clf.fit(X_train, y_train)

# Testar o classificador
accuracy = clf.score(X_test, y_test)

print(f'Acurácia do Classificador MLP: {accuracy}')
```
